```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from scipy.stats import pearsonr
```

```python
columns = ['Name', 'Age', 'Gender', 'FES score', 'Anxiety Score']
all_scores = pd.read_excel('all_scores.xlsx')[columns]
all_scores
```

```python
all_scores.boxplot(column=['Anxiety Score', 'FES score'],by='Gender')
```

```python
all_scores.groupby('Gender').count()
```

```python
all_scores_m = all_scores[all_scores['Gender'] == 'Male']
all_scores_f = all_scores[all_scores['Gender'] == 'Female']

def outlier_removal(df):
  numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
  newdf = df.select_dtypes(include=numerics)

  for x in newdf:
      q3,q1 = np.percentile(df.loc[:,x],[75,25])
      iqr = q3-q1

      max = q3+1.5*iqr
      min = q1-1.5*iqr

      newdf.loc[newdf[x] < min,x] = np.nan
      newdf.loc[newdf[x] > max,x] = np.nan

  return newdf.dropna(axis='rows')

all_scores_m_outlier = outlier_removal(all_scores_m)
all_scores_f_outlier = outlier_removal(all_scores_f)
all_scores_outlier = pd.concat([all_scores_m_outlier, all_scores_f_outlier],ignore_index=True)
```

```python
all_scores_outlier
```

```python
def calculate_corelation(df):
  return pd.DataFrame(pearsonr(df['FES score'],df['Anxiety Score'])).T.rename({0: 'Correlation', 1: 'P-value'},axis=1)

def correlations(dfm, dff, dfcomb):
  return calculate_corelation(dfm), calculate_corelation(dff), calculate_corelation(dfcomb)
```

```python
print(correlations(all_scores_m_outlier, all_scores_f_outlier, all_scores_outlier)[0])
```

```python
sample_size = 100
for i in range(20):
  print(i)
  sample_male = all_scores_m_outlier.sample(n=sample_size, random_state=i)
  sample_female = all_scores_f_outlier.sample(n=sample_size, random_state=i)
  sample_all = pd.concat([sample_male, sample_female],ignore_index=True)
  corrs = correlations(sample_male, sample_female, sample_all)
  # print(corrs[0])
  # print(corrs[1])
  print(corrs[2])
  print('-----------------------------------------------------------------------------------------')
```
