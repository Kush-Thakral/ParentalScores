```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
rng = np.random.default_rng()
s = rng.normal(mu, sigma, 1000)
```

```python
count, bins, _ = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
# desired correlation matrix
cor_matrix = np.array([[1.0, -0.197],
                       [-0.197, 1.0]])

L = np.linalg.cholesky(cor_matrix)

# build some signals that will result in the desired correlation matrix
# X = L.dot(np.random.normal(17,51,(1,100)), np.random.normal(200,292,(1,100))) # the more the sample (1000) the better
# X = L.dot(np.random.normal(0,1,(2,150)))
X = np.vstack((np.random.normal(17,51,(1,100))[0], np.random.normal(292,200,(1,100))[0]))
X = L.dot(X)
# estimate their correlation matrix
print(np.corrcoef(X))
print(X.shape)
print(pearsonr(X[0],X[1]))
plt.scatter(X[0], X[1])
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from scipy.stats import sem

from scipy import stats
from sklearn.model_selection import train_test_split
```

```python
scores = pd.read_csv('trial2.csv').drop('Unnamed: 0',axis=1)
scores.shape
```

```python
pearsonr(scores['anxiety'],scores['FES'])
```

```python
males, females = train_test_split(scores, test_size=0.5)
```

```python
females['anxiety'].mean(), males['anxiety'].mean()
```

```python
male_anxiety = females
female_anxiety = males
```

```python
def metrics(df, column_name = 'anxiety'):
    print(f'mean: {df[column_name].mean()}')
    print(f'median: {df[column_name].median()}')
    print(f'std: {df[column_name].std()}')
    print(f'var: {df[column_name].var()}')
    print(f'skew: {df[column_name].skew()}')
    print(f'kurt: {df[column_name].kurt()}')
```

```python
metrics(male_anxiety)
```

```python
metrics(female_anxiety)
```

```python
stats.ttest_ind(male_anxiety['anxiety'], female_anxiety['anxiety'])
```

```python
stats.ttest_ind(male_anxiety['anxiety'], female_anxiety['anxiety'],equal_var=False)
```
